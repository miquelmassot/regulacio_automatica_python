```python
# System matrices
num = [1]
den = [1, b/m, k/m]
sys = tf(num, den)

sys
```

```python
# Step response for the system
plt.figure(1)
yout, T = step(sys2)
plt.plot(T.T, yout.T)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("y(t) (m)")
plt.show(block=False)
```

```python
# Bode plot for the system
plt.figure(2)
mag, phase, om = bode(sys, logspace(-2, 2), Plot=True)
plt.show(block=False)
```

```python
# Nyquist plot for the system
plt.figure(3)
nyquist(sys, logspace(-2, 2))
plt.show(block=False)
```
